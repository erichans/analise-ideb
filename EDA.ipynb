```python
anos_finais_df = pd.read_excel('divulgacao_anos_finais-escolas-2017.xlsx', header=[4,5,6], na_values='-', skiprows=[7])
```

```python
anos_iniciais_df = pd.read_excel('divulgacao_anos_iniciais-escolas-2017.xlsx', header=[4,5,6], na_values='-', skiprows=[7])
```

```python
anos_iniciais_df.index.names = ['UF']
for column in anos_iniciais_df.columns:
    print(column)
# print(anos_iniciais_df.loc['RO'])
```

```python
anos_finais_df.index.names = ['UF']
for column in anos_finais_df.columns:
    print(column)
```

```python
anos_iniciais_df.sort_values(by=('IDEB\n2017\n(N x P)', 'Unnamed: 81_level_1', 'Unnamed: 81_level_2'), \
                             ascending=False).iloc[:100][['Nome da Escola', 'IDEB\n2017\n(N x P)']] \
            .groupby('UF').count() \
            .sort_values(by=('IDEB\n2017\n(N x P)', 'Unnamed: 81_level_1', 'Unnamed: 81_level_2'), ascending=False)
```

```python
anos_finais_df.sort_values(by=('IDEB\n2017\n(N x P)', 'Unnamed: 74_level_1', 'Unnamed: 74_level_2'), \
                             ascending=False).iloc[:100][['Nome da Escola', 'IDEB\n2017\n(N x P)']] \
            .groupby('UF').count() \
            .sort_values(by=('IDEB\n2017\n(N x P)', 'Unnamed: 74_level_1', 'Unnamed: 74_level_2'), ascending=False)
```

```python
anos_iniciais_top_100 = anos_iniciais_df.sort_values(by=('IDEB\n2017\n(N x P)', 'Unnamed: 81_level_1', 'Unnamed: 81_level_2'), \
                             ascending=False).iloc[:100][['Nome da Escola', 'IDEB\n2017\n(N x P)']]

anos_finais_top_100 = anos_finais_df.sort_values(by=('IDEB\n2017\n(N x P)', 'Unnamed: 74_level_1', 'Unnamed: 74_level_2'), \
                             ascending=False).iloc[:100][['Nome da Escola', 'IDEB\n2017\n(N x P)']]

anos_iniciais_top_100.columns = ['Escola', 'IDEB 2017 (N x P)']
anos_finais_top_100.columns = ['Escola', 'IDEB 2017 (N x P)']
total = anos_iniciais_top_100.append(anos_finais_top_100)

total.sort_values(by='IDEB 2017 (N x P)', ascending=False).iloc[:100] \
    .groupby('UF').count() \
            .sort_values(by='IDEB 2017 (N x P)', ascending=False)


# ('IDEB\n2017\n(N x P)', 'Unnamed: 81_level_1', 'Unnamed: 81_level_2')
# anos_finais_df.sort_values(by=('IDEB\n2017\n(N x P)', 'Unnamed: 74_level_1', 'Unnamed: 74_level_2'), \
#                              ascending=False).iloc[:100][['Nome da Escola', 'IDEB\n2017\n(N x P)']] \
#             .groupby('UF').count() \
#             .sort_values(by=('IDEB\n2017\n(N x P)', 'Unnamed: 74_level_1', 'Unnamed: 74_level_2'), ascending=False)
```
